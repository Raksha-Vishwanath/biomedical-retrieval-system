# BioSeek Diversity Analysis
Visualize diversity metrics by retrieval method.


```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../eval/results/unified/diversity/dense/method_level_diversity.csv')
df
```

```python
plot_df = df.set_index('method')[['clusters_represented', 'cluster_entropy', 'theme_coverage']]
plot_df.plot(kind='bar', figsize=(10, 5), title='Method-Level Diversity Metrics')
plt.tight_layout()
plt.show()
```
