# BioSeek Query-Type Analysis
Load the Stage 6 outputs and plot method performance by bucket.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../results/unified/query_type_analysis/per_bucket_metrics.csv')
df
```

```python
pivot = df.pivot(index='bucket', columns='method', values='ndcg_at_10')
pivot.plot(kind='bar', figsize=(10, 5), title='nDCG@10 by Query Bucket and Method')
plt.ylabel('nDCG@10')
plt.tight_layout()
plt.show()
```

```python
pivot = df.pivot(index='bucket', columns='method', values='precision_at_10')
pivot.plot(kind='bar', figsize=(10, 5), title='Precision@10 by Query Bucket and Method')
plt.ylabel('Precision@10')
plt.tight_layout()
plt.show()
```
